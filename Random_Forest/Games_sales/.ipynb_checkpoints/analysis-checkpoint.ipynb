## Analisis del dataset


## Análisis de Columnas Descartadas

A continuación, se presenta la lista de columnas que no se tomaron en cuenta en el análisis, junto con las razones de su exclusión:

### Columnas Descartadas

1. **img**  
   - **Razón**: No porta un valor significativo para el análisis de popularidad o relevancia de un videojuego. Además, los elementos visuales como imágenes pueden cambiar constantemente, especialmente si el videojuego pertenece a un modelo de servicio en evolución.

2. **console**  
   - **Razón**: Aunque podría ser un buen indicador para analizar dónde se consumen más videojuegos, no aporta directamente a entender la popularidad. Esto se debe a que muchas consolas actuales ofrecen ports a PC u otras plataformas, como el plan de Xbox para convertirse en un publisher más allá del hardware.

3. **publisher**  
   - **Razón**: Aunque puede ser interesante, la popularidad depende más de la comunidad que sigue a los desarrolladores. La interacción con los developers es más directa que con los publishers, por lo que esta columna no resulta prioritaria para este análisis.

4. **critic_score**  
   - **Razón**: Podría tener relevancia si estuviera acompañada de una columna que indicara el total de críticas recibidas. Sin embargo, sin esta información, las calificaciones pueden estar sesgadas si provienen de pocas críticas con alta valoración. Además, no se toma en cuenta la percepción de los consumidores, que es crucial para medir el impacto real. Cabe destacar que el año 2024 marcó una brecha entre la prensa y los gamers (conocida como *GamerGate 2*), reforzando la decisión de no incluir esta columna.

5. **na_sales, jp_sales, pal_sales y other_sales**  
   - **Razón**: Estas columnas podrían ser relevantes para medir el impacto de géneros, títulos o estudios en diferentes regiones. Sin embargo, ya contamos con una columna que resume el total de ventas, haciendo redundantes estos datos específicos por región.

6. **release_date**  
   - **Razón**: Aunque puede ser útil en algunos análisis, su relevancia disminuye en casos de videojuegos que explotan en popularidad años después de su lanzamiento, como *Among Us*. Además, juegos como servicio son atemporales, y títulos icónicos como *Minecraft* refuerzan esta idea.

7. **last_update**  
   - **Razón**: Podría aportar valor si se especificara si se trata de parches, correcciones de bugs, expansiones o DLCs. Sin esta distinción, la columna se considera ruido en este análisis.

---

## Variables Importantes

A continuación, se detallan las variables seleccionadas para el análisis, junto con sus justificaciones:

1. **title**  
   - **Razón**: Sirve principalmente como un apuntador de referencia. Aunque aporta contexto, no tiene un peso significativo en el análisis.

2. **genre**  
   - **Razón**: Algunos jugadores adquieren videojuegos basándose únicamente en su género. Por ejemplo, títulos como *FIFA* o *Call of Duty (COD)* se han convertido en sinónimos de sus propios géneros.

3. **developer**  
   - **Razón**: En la actualidad, los jugadores pueden interactuar y tener cierta influencia en los desarrolladores de sus juegos favoritos. Esto fomenta la creación de comunidades sólidas, haciendo que esta variable sea relevante.

4. **total_sales**  
   - **Razón**: Es la característica más importante del análisis. En ausencia de otras métricas como jugadores activos, impacto mediático a lo largo del tiempo, etc., el rendimiento del juego se medirá en función del dinero generado.  
     Esto es consistente con el mercado actual: un producto puede ser de alta calidad (cumpliendo las expectativas de consumidores y estándares del producto), pero no necesariamente se convierte en un bestseller o alcanza la rentabilidad esperada.

---

## Limpieza del Dataset y Posterior Análisis

### Proceso de Limpieza

1. **Valores Nulos**  
   - Se eliminarán las filas donde `total_sales` sea nulo, ya que esta es la variable clave para el análisis.  
   - En caso de valores nulos en otras columnas, como `developer`, se aplicarán estrategias específicas (por ejemplo, asignar un valor genérico como "Underground 1").

2. **Normalización**  
   - Las ventas están representadas en millones, por lo que no es necesario aplicar una normalización adicional.

3. **Eliminación de Ruido**  
   - Filas con `total_sales` igual a 0 no serán incluidas, ya que no aportan valor y generan ruido en el análisis.

---

### Análisis del Dataset Limpio

Una vez limpio el dataset, se procederá a:  
1. Analizar la densidad de las poblaciones dentro del conjunto de datos, identificando patrones en las ventas por género y desarrollador.  
2. Plantear el etiquetado de juegos en las categorías:
   - **No relevante**  
   - **Relevante**  
   - **Muy relevante**

Este etiquetado permitirá estructurar los datos para análisis posteriores más detallados.

```python
# Importar las librerías necesarias
import pandas as pd
```

```python
# Cargar el dataset original
file_path = './datasets/vgchartz-2024.csv'  # Ruta al archivo
original_df = pd.read_csv(file_path)
```

```python
# Crear una copia filtrada del dataset
filtered_df = original_df.drop(columns=[
    'img', 
    'publisher', 
    'critic_score', 
    'na_sales', 
    'jp_sales', 
    'pal_sales', 
    'other_sales', 
    'release_date', 
    'last_update'
])
```

```python
# Verificar valores nulos en el dataset filtrado
null_info = filtered_df.isnull().sum()
null_percentage = (filtered_df.isnull().sum() / len(filtered_df)) * 100
```

```python
# Mostrar resultados
print("Valores nulos por columna:")
print(null_info)
print("\nPorcentaje de valores nulos por columna:")
print(null_percentage)
```

```python
# Eliminar filas donde total_sales es nulo o igual a 0
filtered_df = filtered_df[filtered_df['total_sales'].notnull() & (filtered_df['total_sales'] > 0)]

# Verificar valores nulos en 'developer' después de la limpieza
null_info_post_cleaning = filtered_df.isnull().sum()
null_percentage_post_cleaning = (filtered_df.isnull().sum() / len(filtered_df)) * 100

# Mostrar resultados
print("Valores nulos por columna después de eliminar 'total_sales' nulos o ceros:")
print(null_info_post_cleaning)
print("\nPorcentaje de valores nulos por columna después de la limpieza:")
print(null_percentage_post_cleaning)
```

```python
# Rellenar valores nulos en 'developer' con nombres genéricos (corrección del aviso)
filtered_df['developer'] = filtered_df['developer'].fillna('Underground 1')

# Contar las repeticiones en 'title'
title_repetition_stats = filtered_df.groupby('title')['console'].count().sort_values(ascending=False)

# Mostrar estadísticas de repeticiones en 'title'
print("Estadísticas de títulos repetidos por número de consolas:")
print(title_repetition_stats)
```

```python
# Filtrar filas con total_sales > 0
filtered_df = filtered_df[filtered_df['total_sales'] > 0]

# Crear una columna auxiliar para encontrar el título más corto en cada grupo
filtered_df['short_title'] = (
    filtered_df.groupby(['genre', 'developer'])['title']
    .transform(lambda x: x.loc[x.str.len().idxmin()])  # Toma el título más corto
)

# Agrupar por el título más corto y sumar total_sales
game_sales = (
    filtered_df.groupby(['short_title', 'genre', 'developer'], as_index=False)
    .agg({'total_sales': 'sum'})
    .rename(columns={'short_title': 'title'})  # Renombrar columna auxiliar
)

# Exportar a un archivo CSV
game_sales.to_csv('./datasets/game_sales.csv', index=False)

print("Nuevo dataset 'game_sales.csv' creado con éxito.")
print("Primeras filas del nuevo dataset:")
print(game_sales.head())
```

```python
import pandas as pd

# Ordenar el dataset por ventas totales
game_sales = filter_df
game_sales = game_sales.sort_values(by='total_sales', ascending=False).reset_index(drop=True)

# Calcular umbrales
percentile_90 = game_sales['total_sales'].quantile(0.9)
top_100_sales = game_sales.iloc[99]['total_sales']  # Ventas del juego en la posición 100

# Asignar etiquetas
def categorize_sales(row):
    if row['total_sales'] > top_100_sales:
        return 'Muy Relevante'
    elif row['total_sales'] > percentile_90:
        return 'Relevante'
    else:
        return 'Irrelevante'

game_sales['category'] = game_sales.apply(categorize_sales, axis=1)

# Verificar distribución
print(game_sales['category'].value_counts())

# Guardar el dataset con categorías
game_sales.to_csv('./datasets/game_sales_categorized.csv', index=False)
```

```python
# Obtener estadísticos por categoría
ranges_by_category = game_sales.groupby('category')['total_sales'].agg(['min', 'max', 'mean', 'median', 'std'])

print(ranges_by_category)
```
